````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Önce verileri görelim
def analyze_oxygen_data(df):
    # Veri tiplerini kontrol et
    print("Veri tipleri:")
    print(df.dtypes)
    
    try:
        # DMO-ÖZEL-İHALE-DEMO dağılımı
        print("\nSatış Türü Dağılımı:")
        print(df['DMO-ÖZEL-İHALE-DEMO'].value_counts())
        
        # Ülkelere göre dağılım
        print("\nÜlkelere Göre Dağılım:")
        print(df['İL'].value_counts())
        
        # Ürün modellerine göre dağılım
        if 'ÜRÜN MODEL' in df.columns:
            print("\nÜrün Modellerine Göre Dağılım:")
            print(df['ÜRÜN MODEL'].value_counts())
        
        # Yıllara göre satış sayıları
        print("\nYıllara Göre Satış Sayıları:")
        yearly_sales = df.groupby('YIL').size()
        print(yearly_sales)
        
        # Görselleştirme
        plt.figure(figsize=(12,6))
        sns.countplot(data=df, x='YIL')
        plt.title('Yıllara Göre Satış Sayıları')
        plt.xticks(rotation=45)
        plt.show()
        
        # Yurtdışı satışların ülkelere göre dağılımı
        export_sales = df[df['DMO-ÖZEL-İHALE-DEMO'] == 'YURTDIŞI']
        plt.figure(figsize=(12,6))
        sns.countplot(data=export_sales, x='İL')
        plt.title('Yurtdışı Satışların Ülkelere Göre Dağılımı')
        plt.xticks(rotation=45)
        plt.show()
        
    except Exception as e:
        print(f"Hata oluştu: {e}")

# Veri setini analiz et
analyze_oxygen_data(combined_df)

# İsterseniz belirli bir yılın verilerini de analiz edebiliriz
# Örnek:
print("\n2023 Yılı Analizi:")
df_2023 = combined_df[combined_df['YIL'] == 2023]
analyze_oxygen_data(df_2023)
````

````python
ÜTS (Ürün Takip Sistemi) özellikle tıbbi cihazlar ve sağlık ürünleri için Türkiye'de kullanılan önemli bir takip sistemidir. 

İşte ÜTS'nin önemi ve detayları:

1. **Temel Amacı:**
- Tıbbi cihazların ve kozmetik ürünlerin güvenli şekilde takibini sağlamak
- Sahte/kaçak ürünlerin önüne geçmek
- Ürünlerin tüm yaşam döngüsünü kayıt altına almak
- Hasta güvenliğini artırmak

2. **Yasal Zorunluluk:**
- Sağlık Bakanlığı tarafından zorunlu tutulan bir sistem
- Tıbbi cihaz üreticileri, ithalatçıları ve distribütörleri için zorunlu
- Oksijen üretim sistemleri tıbbi cihaz sınıfında olduğu için ÜTS kaydı zorunlu

3. **ÜTS Numarasının İçeriği:**
- Her ürüne özgü benzersiz bir kimlik numarası
- Ürünün tekil olarak takibini sağlar
- Barkod sistemiyle entegre çalışır
- 13 haneli numerik bir koddur

4. **Sağladığı Faydalar:**
- **Ürün Güvenliği:**
  - Ürünün orijinalliğinin doğrulanması
  - Kalite standartlarına uygunluğun takibi
  
- **İzlenebilirlik:**
  - Ürünün üretimden son kullanıcıya kadar takibi
  - Servis ve bakım geçmişinin kaydı
  - Oluşabilecek sorunların hızlı tespiti

- **Piyasa Gözetimi:**
  - Piyasadaki ürünlerin kontrolü
  - Uygunsuz ürünlerin tespiti
  - Geri çağırma işlemlerinin yönetimi

5. **Sistemdeki Bilgiler:**
- Ürün teknik özellikleri
- EC sertifikası bilgileri
- Üretici/İthalatçı bilgileri
- Kullanım kılavuzu
- Teknik servis bilgileri

6. **Neden Veri Setinde Önemli:**
- **Yasal Uyumluluk:** 
  - Tüm satışların ÜTS'ye kaydedilmesi zorunlu
  - Denetimlerde kontrol edilen bir alan
  
- **Ürün Takibi:**
  - Garanti süreçlerinin yönetimi
  - Servis geçmişinin takibi
  - Kalite kontrol süreçleri

- **Müşteri Güvencesi:**
  - Ürünün resmi olarak onaylı olduğunu gösterir
  - Servis ve yedek parça güvencesi sağlar

7. **Eksik Olmasının Riskleri:**
- Yasal yaptırımlar
- Ürün satışının engellenmesi
- Müşteri güveninin sarsılması
- Servis ve bakım süreçlerinde zorluklar

8. **Öneriler:**
- Veri setinde ÜTS numarası alanının boş bırakılmaması
- Her ürün için ÜTS kaydının tamamlanması
- ÜTS bilgilerinin düzenli güncellenmesi
- ÜTS entegrasyonunun ERP sistemine dahil edilmesi

9. **Raporlamada Kullanımı:**
- Ürün sertifikasyon durumu takibi
- Yasal uyumluluk raporları
- Kalite kontrol süreçleri
- Servis ve bakım planlaması

Bu nedenlerden dolayı ÜTS numarası, özellikle 2024 ve sonrası veriler için kritik öneme sahip bir alan olup, veri setinde muhafaza edilmesi ve tüm ürünler için doldurulması gereken bir sütundur.
````
